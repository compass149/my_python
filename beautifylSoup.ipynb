```python
import urllib.request as req
base='https://ko.wikisource.org'
url=base+'/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC'
res=req.urlopen(url)
soup=BeautifulSoup(res, 'html.parser')
li_list=soup.select('#mw-content-text > div.mw-parser-output li')

for li in li_list:
    main_a=li.a
    print(main_a.text)
    sub_li_list=li.select('ul>li')
    if len(sub_li_list) > 0:
        for sub_li in sub_li_list:
            a_tag=sub_li.a
            print("-", a_tag.text)
            url=base+a_tag['href']
            res=req.urlopen(url)
            soup=BeautifulSoup(res,'html.parser')
            content=soup.select_one('#mw-content-text >div.mw-parser-output p')
            print(content.text)
    else:
        url=base+main_a['href']
        res=req.urlopen(url)
        soup=BeautifulSoup(res,'html.parser')
        #print(soup)
        content1=soup.select_one('div.mw-parser-output')
        p_list=content1.find_all('p')
        for p in p_list:
            print(p.text)
            
```

```python
from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO
from io import open
from urllib.request import urlopen
```

```python
def readPDF(pdfFile):
    rsrcmgr=PDFResourceManager()
    retstr=StringIO()
    laparams=LAParams()
    device=TextConverter(rsrcmgr,retstr, laparams=laparams)
    process_pdf(rsrcmgr,device,pdfFile)
    device.close()
    content=retstr.getvalue()
    retstr.close()
    return content
pdfFile=urlopen("http://www.pythonscraping.com/pages/warandpeace/chapter1.pdf")
outputString=readPDF(pdfFile)
#print(outputString)
pdfFile.close()
with open("C:/data/result.txt",'w') as f:
    f.write(outputString)
    print("저장되었습니다.")
```

```python
#데이터 수집
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs

def getLinks(page):
    html=urlopen("https://store.kyobobook.co.kr/bestseller/total/weekly")
    print(html)
    obj=bs(html,'html.parser')
    all_text=obj.find_all({"class":"font-weight-bold text-black"})
    data_list=[]
    for s1 in all_text:
        text=s1.get_text()
        textarr=text.replace('신고','').replace('\t','').replace('\n\n\n\n','').split("\n\n")
        textarr[0] = textarr[0].strip('\n')
        textarr[1]=int(textarr[1][textarr[1].find('중')+1:])
        if len(textarr)<3:
            textarr.append('-')
        data_list.append(tuple(textarr))
        
    return data_list
```
