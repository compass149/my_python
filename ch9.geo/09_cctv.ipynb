```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
cctv = pd.read_csv('../data/부산광역시 부산진구_방범용CCTV_20241010.csv', encoding='CP949')
cctv
```

```python
popups = list(cctv.iloc[:,1])
popups
```

```python
counts = list(cctv.iloc[:,4])
counts
```

```python
lats= list(cctv.iloc[:,9])
longs = list(cctv.iloc[:,10])
print(lats)
print(longs)

lat_longs = np.array([lats, longs]).T
lat_longs
```

```python
print(len(popups), len(counts), lat_longs.shape)
```

```python
popups1, lat_longs1 = [], []
for i in range(len(counts)):
    if counts[i] > 0:
        popups1.append(popups[i])
        lat_longs1.append(lat_longs[i])
        
print(len(popups1), len(lat_longs1))

lat_longs_arr = np.array(lat_longs1)
lat_longs_arr
```

```python
import folium
from folium import plugins
import os

# 예제 데이터
lat_longs_arr = [
    [35.1685836, 129.0379501],
    [35.1695836, 129.0389501],
    [35.1705836, 129.0399501]
]
popups1 = ['Location 1', 'Location 2', 'Location 3']

# 지도 생성
m = folium.Map(location=[35.1685836, 129.0379501], zoom_start=15)

# MarkerCluster 추가
plugins.MarkerCluster(lat_longs_arr, popups=popups1).add_to(m)

# 지도 저장
m.save(os.path.join('../data', 'cctv.html'))

# 지도 표시 (Jupyter Notebook에서)
m
```
