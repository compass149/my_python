```python
from folium import plugins
import os
N = 100
data = np.array(
    [
        np.random.uniform(low=35, high=38, size=N), # Random latitudes.
        np.random.uniform(low=126, high=129, size=N), # Random longitudes.
    ]).T
popups = [str(i) for i in range(N)] # Popups texts are simple numbers.
m = folium.Map([37.566345, 126.977893], zoom_start=8)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m.save(os.path.join('../../data', 'plugins_1.html'))
m
```

```python
import folium
import pandas as pd
import requests

# (예시) 주별 실업률 가짜 데이터
data = {
    'state_id': ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA'],  # 주 약자
    'unemp': [3.2, 4.3, 4.0, 3.9, 5.1, 2.8, 3.4, 4.5, 2.9, 4.2]               # (예시) 실업률
}
df = pd.DataFrame(data)

# 지도 생성
m = folium.Map(location=[43, -100], zoom_start=4)

# GeoJSON 불러오기
url = "https://raw.githubusercontent.com/python-visualization/folium-example-data/main/us-states.json"
us_states = requests.get(url).json()

# Choropleth (주별 실업률 시각화)
folium.Choropleth(
    geo_data=us_states,
    data=df,
    columns=['state_id', 'unemp'],        # df에서 [키, 값]에 해당하는 컬럼
    key_on='feature.id',                  # us-states.json에서 각 주의 id와 df의 state_id를 매칭
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Unemployment Rate (%)'
).add_to(m)

m
```

```python
cb = pd.read_csv('../../data/CoffeeBean.csv', encoding='cp949', index_col=0, header=0)
cb
```

```python
addr = pd.DataFrame(cb.address.apply(lambda v: v.split()[:2]).tolist(), columns=['시도', '구군'])
addr
```

```python
addr.시도.unique()
```

```python
addr_aliases = {'서울':'서울특별시', '서울시':'서울특별시'}
addr['시도'] = addr['시도'].apply(lambda v: addr_aliases.get(v, v))
addr
```

```python
addr2 = addr.loc[addr.시도 == '서울특별시',:]
addr2
```

```python
addr_group = addr2.groupby('구군')['시도'].count()
#addr_group = addr_.group.rename(columns={"시도":'매장수'})
df=pd.DataFrame({'gugun':addr_group.index, 'count':addr_group.values})
#addr_count.columns = ['구군', '매장수']
df
```

```python
m = folium.Map(
    location=[37.566345, 126.977893],
    zoom_start=11,
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)
m
```

```python
folium.Choropleth(
    geo_data=seoul_geo,
    data=df,
    columns=['gugun', 'count'],
    key_on='properties.name',
    fill_color='RdYlGn',
    bins=[0,1,2,3,4,5,10,20,30,57],
    fill_opacity=0.3,
    line_opacity=0.1,
    legend_name='지역구별 커피 업종 수',
    highlight=True
).add_to(m)
m
```
